## Análisis de la relación entre estatura y desempeño en arqueros

```python
df_gk = df[df['position'] == 'GK']
```

## Categorías de altura

```python
def categorize_height(height):
    if height < 185:
        return "Menos de 185 cm"
    elif 185 <= height <= 192:
        return "Entre 185 y 192 cm"
    else:
        return "Más de 192 cm"

df_gk["height_category"] = df_gk["height_cm"].apply(categorize_height)
```

## Boxplot de overall según categorías de altura

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x="height_category", y="overall", data=df_gk, palette="pastel")
plt.title("Desempeño (Overall) de los arqueros según estatura")
plt.xlabel("Categoría de Estatura")
plt.ylabel("Overall")
plt.show()
```

## Promedio de overall por categoría de altura

```python
mean_overall_by_height = df_gk.groupby("height_category")["overall"].mean()
print("\nPromedio de Overall por Categoría de Estatura:")
print(mean_overall_by_height)
```

## **Respuesta**

- Menos de 185 cm

Tiene una mediana de desempeño más baja en comparación con los otros dos grupos.
La dispersión de los datos es amplia, con valores atípicos en la parte superior.
El rango intercuartílico (IQR) es más reducido, lo que indica menos variabilidad en comparación con las otras categorías.

- Entre 185 y 192 cm

Presenta una mediana similar al grupo de "Más de 192 cm".
La dispersión de los datos es amplia, y se observan valores atípicos por encima de los 80 puntos de desempeño.
El rango intercuartílico es similar al de los arqueros de más de 192 cm.

- Más de 192 cm

Tiene una mediana de desempeño ligeramente más alta.
También hay valores atípicos en la parte superior.
Aunque la variabilidad es parecida a la categoría intermedia, su distribución parece estar un poco más centrada.

- Conclusiones del análisis descriptivo

Tendencia de desempeño: Los arqueros más altos tienden a tener un mejor desempeño promedio en comparación con los más bajos.

Variabilidad en el desempeño: Existe variabilidad en todas las categorías, lo que indica que la estatura no es el único factor determinante del desempeño.

Valores atípicos: Se observan valores atípicos en todas las categorías, lo que sugiere que algunos arqueros excepcionalmente buenos o malos están presentes en cada grupo.
